```
# 도전문제 5.2: pairplot()의 kind 사용하기(난이도: 중)
# 시본 라이브러리의 pairplot() 함수에 kind='hist' 키워드 인자를 추가하여 왼쪽 그림과 같이 표시하여라.
# 다음으로 kind='kde'를 사용하여 쌍 플롯을 그려 보아라.
import seaborn as sns
tips = sns.load_dataset('tips')
sns.pairplot(tips,kind='hist')
sns.use_style('')
```

```
# 5.1 시본의 lmplot() 함수를 사용하여 다음과 같이 tip 데이터를 시각화하여라. 이 데이터는 다음과 같이 smoker 값이
# Yes/No에 따라서 서로 다른 선형 회귀 직선이 하나의 그래프에 나타난다.
import seaborn as sns
sns.set_theme(style='darkgrid')
tips = sns.load_dataset('tips')
print(tips)
sns.lmplot(data=tips,x="total_bill",y="tip",hue="smoker")
```
