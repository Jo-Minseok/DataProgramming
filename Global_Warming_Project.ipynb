```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf

tf.random.set_seed(777)
step = 10

def Data_Set(data):
  X = []
  Y = []
  for i in range(len(data) - step):
    end_index = i + step
    X.append(data[i:end_index])
    Y.append(data[end_index])
  return (np.array(X), np.array(Y))

def Model_Set():
  model = Sequential()
  model.add(LSTM(step, return_sequences = True, input_shape = (step, 1)))
  model.add(LSTM(step * 2, return_sequences = False))
  model.add(Dense(1, activation = 'linear'))
  model.compile(optimizer = 'adam', loss = 'mse')
  return model

data = MinMaxScaler().fit_transform(np.array(tp['평균기온(°C)']).reshape(-1, 1))
x, y = Data_Set(data)
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, shuffle = False)
x_train = x_train.reshape(x_train.shape[0], step, 1)
x_test = x_test.reshape(x_test.shape[0], step, 1)

model = Model_Set()
model.fit(x_train, y_train, epochs = 20, batch_size = 2)
pred = model.predict(x_test)

plt.plot(y_train)
plt.title("실데이터를 통한 온도 예측")
plt.scatter(np.arange(len(y_train), len(y_train) + len(y_test)), y_test, c = 'r', label = "실제 데이터")
plt.scatter(np.arange(len(y_train), len(y_train) + len(pred)), pred, c = 'b', label = "예측 값")
plt.legend()
```

```python
pr = data

for i in range(100):
  hat = model.predict([list(pr[-step:])])
  pr = np.append(pr, hat)

print(pr)
plt.plot(pr)
```

```python
from sklearn.linear_model import LinearRegression

def Pred_Temp(data, end_year = 2100):
  x = np.array(data.index).reshape(-1, 1)
  y = data
  lr = LinearRegression()
  lr.fit(x, y)
  y_predict = lr.predict(x)

  diff = np.array([])
  for i in range(tp.index[0], tp.index[-1] + 1):
    diff = np.append(diff, y.loc[i] - y_predict[i - tp.index[0]])

  pred = lr.predict(np.arange(tp.index[0], end_year).reshape(-1, 1))
  plt.plot(np.arange(tp.index[0], tp.index[-1] + 1), y, label = data.name)
  plt.plot(np.arange(tp.index[0], end_year), pred, 'g', label = '선형회귀 직선')
  plt.plot(np.arange(tp.index[0], end_year), pred + diff.max(), 'r', label = '선형회귀 직선(상한값)')
  plt.xlabel('년도')
  plt.ylabel('기온(°C)')
  plt.title(data.name + ' 선형회귀 직선')
  plt.legend()
```

```python
Pred_Temp(tp['평균최고기온(°C)'])
```

```python
from statsmodels.tsa.arima_model import ARIMA

def auto_arima(data, order, sort = 'AIC'):
  order_list = []
  aic_list = []
  bic_list = []
  for p in range(order[0]):
    for d in range(order[1]):
      for q in range(order[2]):
        model = ARIMA(data, order = (p, d, q))
        try:
          model_fit = model.fit()
          c_order = f'p{p} d{d} q{q}'
          aic = model_fit.aic
          bic = model_fit.bic
          order_list.append(c_order)
          aic_list.append(aic)
          bic_list.append(bic)
        except:
          pass

  result_df = pd.DataFrame(list(zip(order_list, aic_list)), columns = ['order', sort])
  result_df.sort_values(sort, inplace = True)
  return result_df


auto_arima(tp['평균기온(°C)'], [3, 3, 3])
```
