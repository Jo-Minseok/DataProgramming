```
#도전문제 4.1(난이도: 중)
# 다음과 같은 그림을 subplots 함수와 GridSpec 클래스를 사용하여 화면에 그리도록 하여라. 이 화면을 구성하기 위해서는 5개의 격자 공간이 필요하며 산점도와 막대 그래프,
# 이미지 자료 값은 임의의 값을 사용하라.
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(3,3)
grid = plt.GridSpec(3,3,wspace=0.4,hspace=0.3)

X = np.random.uniform(-2,4,100)
Y = np.random.randn(100)
plt.subplot(grid[0,:3]).scatter(X,Y)

Z = np.random.uniform(0,5,([10,5]))
plt.subplot(grid[1:,2]).imshow(Z)

X = [1,2,3,4,5]
plt.subplot(grid[1,0:2]).plot(X,'r--')

X = np.linspace(0,10,100)
Y = np.cos(X)
plt.subplot(grid[2,0]).plot(X,Y,'g-')

X = np.arange(10)
Y = np.random.randint(0,11,10)
plt.subplot(grid[2,1]).bar(X,Y)
```

```
# 도전문제 4.2(난이도: 중)
# barh()는 수평 막대 그림을 그리는 함수이다. 위의 제주도 관광객 방문 그래프 그림을 barh() 함수를 사용하여 다음과 같은 수평 막대 그림으로 그려보도록 하자.
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(3)
year = ['2010','2011','2012']
domestic = [6801,7695,8010]
foreign = [777,1046,1681]

plt.barh(x,domestic,height=0.25)
plt.barh(x-0.3,foreign,height=0.25)
plt.yticks(x,year)
```

```
# 도전문제 4.3(난이도: 상)
# 오른쪽 그림과 같이 100개의 난수를 가진 데이터 10종을 만들어 각각의 데이터 범위를 가시화하는 코드를 작성해보라. 난수를 생성하는 일은 넘파이의 radnom 모듈이 가지고 있는 randn(do, d1)
# 을 사용하면 될 것이다.
# 이때 생성되는 데이터는 넘파이 배열이므로 차원을 어떻게 결정해야 하는지 유의해서 숫자 d0와 d1을 결정해야 할 것이다. 결과는 생성된 난수에 따라 오른쪽 그림과 조금 다를 것이다.

import numpy as np
import matplotlib.pyplot as plt

data1=np.random.randn(100)
data2=np.random.randn(100)
data3=np.random.randn(100)
data4=np.random.randn(100)
data5=np.random.randn(100)
data6=np.random.randn(100)
data7=np.random.randn(100)
data8=np.random.randn(100)
data9=np.random.randn(100)
data10=np.random.randn(100)

plt.boxplot([data1,data2,data3,data4,data5,data6,data7,data8,data9,data10])
```

```
#4.1
#1) 대표적인 주기함수인 사인 함수를 사용하여 다음 그림과 같은 히트맵을 그려보도록 하자. 격자의 크기는 10x10 이며, 사인 함수의 입력값의 구간은 0에서 파이 사이이다.
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,np.pi,100).reshape(10,10)
y = np.sin(x)
plt.imshow((y+np.flip(y,axis=0)+y.T+np.flip(y.T,axis=1))/4)
plt.colorbar()
```

```
#2) 사인 함수를 사용하여 100 x 100 크기의 격자 공간에 다음과 같은 패턴이 나타나도록 위의 코드를 수정하여라.
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,np.pi,10000)
y = np.sin(x).reshape(100,100)
plt.imshow(y.T)
```

```
# 4.2 난수를 1000개 발생시켜 임의의 2차원 좌표를 생성해 그려보자. 이때 좌표의 x와 y 값은 표준 정규 분포를 따르도록 할 것이다. 그러면 생성된 좌표는 원점 (0,0)에 밀집한 모양을
# 가질 것이다.
# 1) scatter() 함수 내에 불투명도를 의미하는 alpha 키워드 매개변수를 0.01로 두고 그림을 그려보도록 하자.
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x,y,alpha=0.09,color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random scatter plot')
```

```
# 2) alpha 키워드 매개변수를 0.1로 두고, 색상을 'magenta'로 두어 심홍색 점이 진하게 보이도록 그려보자.
# 힌트: 이 alpha 값은 0에서 1 사이 값을 가지며, 0이면 투명, 1이면 불투명이다.
x=np.random.randn(1000)
y=np.random.randn(1000)
plt.scatter(x,y,alpha=0.1,color='magenta')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Random scatter plot')
```

```
# 4.3 정규 분포를 가지는 10,000개의 난수를 생성하여 다음과 같은 f1, f2 다차원 배열을 만들어 보자. 이것을 이용하여 아래와 같은 히스토그램을 생성하여라.
# 힌트) pyplot의 subplot() 함수를 이용하도록 하자
# f1 = np.random.normal(loc=0, scale=1, size=10000)
# f2 = np.random.normal(loc=3, scale=.5, size=10000)
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,2)
grid = plt.GridSpec(2,2)

f1 = np.random.normal(loc=0, scale=1, size=10000)
f2 = np.random.normal(loc=3, scale=.5, size=10000)
plt.subplot(grid[0,0]).hist(f1,color='r',bins=200)
plt.title('Red histogram')
plt.subplot(grid[0,1]).hist(f2,color='b',bins=200)
plt.title('Blude histogram')
plt.subplot(grid[1,0:]).hist(f1,bins=200,color='r',alpha=.75,label='loc = 0, scale = 1')
plt.subplot(grid[1,0:]).hist(f2,color='b',bins=200,alpha=.75,label='loc = 3, scale = .5')
plt.legend()
```
